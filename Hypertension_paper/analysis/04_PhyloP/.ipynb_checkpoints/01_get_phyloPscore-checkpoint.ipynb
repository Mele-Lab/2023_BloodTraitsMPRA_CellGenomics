### functions 

```python
def diff_sig_status(row):
    if row.CM_status == "significant effect" and row.VSMC_status == "significant effect":
        return "sig_both"
    elif row.CM_status == "significant effect" and row.VSMC_status == "no effect":
        return "sig_CM"
    elif row.CM_status == "no effect" and row.VSMC_status == "significant effect":
        return "sig_VSMC"
    else:
        return "not_sig_both"
```

## Read files 

```python
sentinel_snps = pd.read_csv(sentinel_snps_f, sep = '\t')
sentinel_snps = sentinel_snps[sentinel_snps['Trait'].isin(['DBP','PP','SBP'])]
print(sentinel_snps.shape)
sentinel_snps.head()
```

```python
phylop_hg19 = pd.read_csv(phylop_hg19_f_a, sep='\t')
phylop_hg19.head()
```

```python
CM_results.shape
```

```python
VSMC_results = pd.read_csv(VSMC_results_f, sep = '\t')
VSMC_results.head()
```

## 2. Merge PhyloP scores with SNP info 

```python
# for window
sentinel_snps[['#chrom','pos','REF','ALT']] = sentinel_snps.change.str.split(":",expand=True)
sentinel_snps['start'] = sentinel_snps['pos'].astype(int)+1
sentinel_snps['end'] = sentinel_snps['pos'].astype(int)+11
sentinel_snps['#chrom'] = 'chr'+sentinel_snps['#chrom']
sentinel_snps.sample(5)
```

```python
sentinel_snps_merged = sentinel_snps.merge(phylop_hg19, how='left')
```

```python
sentinel_snps_merged.drop_duplicates(inplace=True)
print(sentinel_snps_merged.shape)
sentinel_snps_merged.sample(5)
```

```python
sentinel_snps_merged.to_csv('PhyloP_scores_summary_tile_46_placenta_window_a.txt', sep='\t', index=False)
```

## 3. Merge with CM and VSMC results 

```python
sentinel_snps_merged = phylop_hg19
sentinel_snps_merged.sample(5)
```

```python
sentinel_snps_merged_res = sentinel_snps_merged.merge(CM_results, how='left', on=['snp']).merge(VSMC_results, how='left', on=['snp'], suffixes=('_CM','_VSMC'))
sentinel_snps_merged_res.head()
```

```python
sentinel_snps_merged_res.shape
```

```python
def diff_sig_status(row):
    if row.fdr_comp_CM < 0.05 and row.fdr_comp_VSMC < 0.05:
        return "sig_both"
    elif row.fdr_comp_CM < 0.05:
        return "sig_CM"
    elif row.fdr_comp_VSMC < 0.05:
        return "sig_VSMC"
    else:
        return "not_sig_both"
```

```python
sentinel_snps_merged_res['sig_status'] = sentinel_snps_merged_res.apply(diff_sig_status, axis=1)
sentinel_snps_merged_res.sig_status.value_counts()
```

```python
sentinel_snps_merged_res.sample(5)
```

## 4. plot

```python
sns.set(**PAPER_PRESET)
fontsize = PAPER_FONTSIZE
```

```python
order1 = ["not_sig_both", "sig_VSMC", "sig_CM","sig_both"]
pal2 = {"not_sig_both": "darkgray", "sig_VSMC": sns.color_palette("Set2")[1], 
        "sig_CM": sns.color_palette("Set2")[0], "sig_both": sns.color_palette("Set2")[2]}


tmp = sentinel_snps_merged_res.copy()

fig, ax = plt.subplots(figsize=(2.5, 2), nrows=1, ncols=1)

sns.boxplot(data=tmp, x="sig_status", y=(tmp["mean"]), order=order1, palette=pal2, 
            flierprops = dict(marker='o', markersize=5), ax=ax)
mimic_r_boxplot(ax)
ax.set_xticklabels(["not sig", "sig in VSMC", "sig in CM", "sig in both"], rotation=50, 
                   ha="right", va="top")
ax.set_ylabel("mean(PhyloPScore)")
ax.set_title("")
ax.set_xlabel("")

for i, label in enumerate(order1):
    n = len(tmp[tmp["sig_status"] == label])
    ax.annotate(str(n), xy=(i, -8), xycoords="data", xytext=(0, 0), 
                textcoords="offset pixels", ha='center', va='top', 
                color=pal2[label], size=fontsize)

ax.set_ylim((-10, 11))
#ax.set_ylim((-1.5, 4))


# calc p-vals b/w dists
dist1 = np.asarray(tmp[tmp["sig_status"] == "not_sig_both"]["mean"])
dist2 = np.asarray(tmp[tmp["sig_status"] == "sig_VSMC"]["mean"])
dist3 = np.asarray(tmp[tmp["sig_status"] == "sig_CM"]["mean"])
dist4 = np.asarray(tmp[tmp["sig_status"] == "sig_both"]["mean"])

dist1 = dist1[~np.isnan(dist1)]
dist2 = dist2[~np.isnan(dist2)]
dist3 = dist3[~np.isnan(dist3)]
dist4 = dist4[~np.isnan(dist4)]

u, pval1 = stats.mannwhitneyu(dist1, dist2, alternative="two-sided", use_continuity=False)
u, pval2 = stats.mannwhitneyu(dist1, dist3, alternative="two-sided", use_continuity=False)
u, pval3 = stats.mannwhitneyu(dist1, dist4, alternative="two-sided", use_continuity=False)
#u, pval3 = stats.mannwhitneyu(dist1, dist4, alternative="two-sided", use_continuity=False)
print(pval1)
print(pval2)
print(pval3)
print(stats.mannwhitneyu(dist1, dist4, alternative="two-sided", use_continuity=False))
print(np.mean(dist1))
print(np.mean(dist4))

#annotate_pval(ax, 0.2, 0.8, 5.3, 0, 5.3, pval1, fontsize)
ax.plot([0.2, 0.2, 0.8, 0.8], [2.5, 2.5+0, 2.5+0, 2.5], c="black", lw=0.5)
if pval1 < 0.001:
    ax.text((0.2+0.8)*.5, 2.5, '**', ha='center', va='bottom', color="black", size=fontsize)
elif pval1 < 0.05:
    ax.text((0.2+0.8)*.5, 2.5, '*', ha='center', va='bottom', color="black", size=fontsize)
else:
    ax.text((0.2+0.8)*.5, 2.5, 'n.s.', ha='center', va='bottom', color="black", size=fontsize)
    
ax.plot([0.2, 0.2, 1.8, 1.8], [5, 5+0, 5+0,5], c="black", lw=0.5)
if pval2 < 0.001:
    ax.text((0.2+1.8)*.5, 5, '**', ha='center', va='bottom', color="black", size=fontsize)
elif pval2 < 0.05:
    ax.text((0.2+1.8)*.5, 5, '*', ha='center', va='bottom', color="black", size=fontsize)
else:
    ax.text((0.2+1.8)*.5, 5, 'n.s.', ha='center', va='bottom', color="black", size=fontsize)
    
ax.plot([0.2, 0.2, 2.8, 2.8], [7, 7+0, 7+0, 7], c="black", lw=0.5)
if pval3 < 0.001:
    ax.text((0.2+2.8)*.5, 7, '**', ha='center', va='bottom', color="black", size=fontsize)
elif pval3 < 0.05:
    ax.text((0.2+2.8)*.5, 7, '*', ha='center', va='bottom', color="black", size=fontsize)
else:
    ax.text((0.2+2.8)*.5, 7, 'n.s.', ha='center', va='bottom', color="black", size=fontsize)

#annotate_pval(ax, 0.2, 1.8, 7.3, 0, 7.3, pval2, fontsize)
#annotate_pval(ax, 0.2, 2.8, 10.8, 0, 10.8, pval3, fontsize)

#fig.savefig("FigX_PhyloP.2.snp.svg", dpi="figure", bbox_inches="tight")
plt.show()
```
