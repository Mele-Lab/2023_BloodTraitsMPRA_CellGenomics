## 3. find appropriate FDR cutoffs for each model

choose FDR cutoff as the one that calls < 10% of controls as significant

```python
comp_ctrls = comp[comp["index"].str.contains("SAMP_PAIR")]
comp_ctrls.sample(5)
```

```python
print(len(comp_ctrls))
```

make plots to show different #s of controls called as "significant" at alpha < 0.05 in each model

```python
n_sig = len(comp_ctrls[comp_ctrls["fdr_comp"] < 0.05])
n_sig
```

```python
n_sig_models = {}
for model, df, fdr in zip(["comparison"],
                          [comp_ctrls],
                          ["fdr_comp"]):
    n_sig = len(df[df[fdr] < 0.05])
    n_sig_models[model] = [n_sig]
n_sig_models = pd.DataFrame.from_dict(n_sig_models, orient="index").reset_index()
n_sig_models.columns = ["model", "n_sig"]
n_sig_models
```

```python
fig, ax = plt.subplots(figsize=(2.2, 1.5), nrows=1, ncols=1)

sns.barplot(data=n_sig_models, x="model", y="n_sig", color="darkgray", ax=ax)
ax.set_xlabel("")
ax.set_ylabel("number of controls significant\nat MPRAnalyze q-value < 0.05")
_ = ax.set_xticklabels(n_sig_models["model"], rotation=50, ha="right", va="top")

ax.set_title("MPRAnalyze q-value cut-off")
ax.set_ylim((0, 126))
fig.savefig("Fig_controls_sig.new_back.svg", dpi="figure", bbox_inches="tight")
```

```python
COMP_THRESH = np.percentile(comp_ctrls["fdr_comp"], 5)
COMP_THRESH
```

```python
n_sig_models = {}
for model, df, fdr, thresh in zip(["comparison"],
                                  [comp_ctrls],
                                  ["fdr_comp"],
                                  [COMP_THRESH]):
    n_sig = len(df[df[fdr] < thresh])
    n_sig_models[model] = [n_sig]
n_sig_models = pd.DataFrame.from_dict(n_sig_models, orient="index").reset_index()
n_sig_models.columns = ["model", "n_sig"]
n_sig_models
```

```python
fig, ax = plt.subplots(figsize=(2.2, 1.5), nrows=1, ncols=1)

sns.barplot(data=n_sig_models, x="model", y="n_sig", color="darkgray", ax=ax)
ax.set_xlabel("")
ax.set_ylabel("number of controls significant\nat empirical FDR < 0.002")
_ = ax.set_xticklabels(n_sig_models["model"], rotation=50, ha="right", va="top")

ax.set_title("empirical FDR cut-off")
ax.set_ylim((0, 100))
fig.savefig("Fig_controls_pos_empirical.new_back.svg", dpi="figure", bbox_inches="tight")
```
