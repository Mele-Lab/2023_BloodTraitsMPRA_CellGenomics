## 6.1. pair positive controls together to serve as negative controls
I'm going to downsample each positive tile to 25 and 25 barcodes to serve as negative controls, as they come from the same sequence, no changes are expected

```python
ctrl_ids = rna_counts_VSMC[rna_counts_VSMC['tile_type'].str.contains('CONTROL')].element.unique()
ctrl_ids[0:5]
```

```python
dna_counts_VSMC.head()
```

```python
for i, seq in enumerate(ctrl_ids):
    ctrl_ids[i] = ctrl_ids[i] + '__seq' + str(i)
ctrl_ids[0:5]
```

```python
ctrl_seqs = list([x.split("__")[0] for x in ctrl_ids])
samp_ids = set([x.split("__")[1] for x in ctrl_ids])
```

```python
all_samp_id_pairs = list(itertools.chain.from_iterable(itertools.repeat(x, 3) for x in ctrl_seqs))
all_samp_id_pairs_str = ["%s__%s" % (x, x) for x in all_samp_id_pairs]
all_samp_id_pairs_str[0:5]
```

```python
neg_ctrls_dna = pd.DataFrame()
neg_ctrls_rna = pd.DataFrame()
comp_id_list = []


for i, seq in enumerate(ctrl_seqs):
    print("ctrl #: %s" % (i+1))
    
    for j, samp_id_pair in enumerate(all_samp_id_pairs_str[0:3]):
        if j % 50 == 0:
            print("...samp pair #: %s" % (j+1))
        
        #samp1 = samp_id_pair.split("__")[0] # arbitrarily call 'snp' seq
        #samp2 = samp_id_pair.split("__")[1] # arbitrarily call 'wt' seq
        
        snp_sub_dna = dna_counts_VSMC[dna_counts_VSMC["element"] == seq]
        wt_sub_dna = dna_counts_VSMC[dna_counts_VSMC["element"] == seq]

        snp_sub_rna = rna_counts_VSMC[rna_counts_VSMC["element"] == seq]
        wt_sub_rna = rna_counts_VSMC[rna_counts_VSMC["element"] == seq]

        # re-name columns w/ species name
        snp_dna_cols = ["element"]
        wt_dna_cols = ["element"]
        snp_rna_cols = ["element"]
        wt_rna_cols = ["element"]

        columns_snp =  snp_sub_dna.drop(["element","tile_type",'tile_id'], axis=1, inplace=False)
        columns_wt =  wt_sub_dna.drop(["element","tile_type",'tile_id'], axis=1, inplace=False)

        random_snp = np.random.choice(columns_snp.columns,size=25, replace=False)
        random_wt = np.random.choice(columns_wt.columns,size=25, replace=False) 

        barcodes_snp = list([x.split("__")[1] for x in random_snp])
        barcodes_wt = list([x.split("__")[1] for x in random_wt])

        columns_snp_rna =  snp_sub_rna.drop(["element","tile_type",'tile_id'], axis=1, inplace=False)
        columns_wt_rna =  wt_sub_rna.drop(["element","tile_type",'tile_id'], axis=1, inplace=False)

        ## get rna columns to match barcodes
        snp_rna_cols_df = pd.DataFrame(columns_snp_rna.columns)
        wt_rna_cols_df = pd.DataFrame(columns_wt_rna.columns)

        snp_rna_cols_df.columns = ['column']
        wt_rna_cols_df.columns = ['column']

        new = snp_rna_cols_df["column"].str.split("__", n = 1, expand = True) 
        new_wt = wt_rna_cols_df["column"].str.split("__", n = 1, expand = True) 

        snp_rna_cols_df["barc"]= new[1] 
        wt_rna_cols_df["barc"]= new[1]

        ## subset barcodes 
        snp_rna_cols_df = snp_rna_cols_df[snp_rna_cols_df['barc'].isin(barcodes_snp)]
        wt_rna_cols_df = wt_rna_cols_df[wt_rna_cols_df['barc'].isin(barcodes_wt)]  

        cols_snp = ["element"]+ list(random_snp)
        cols_wt = ["element"]+ list(random_wt)
        
        snp_sub_dna = dna_counts_VSMC[dna_counts_VSMC["element"] == seq]
        snp_sub_dna = snp_sub_dna[cols_snp]
        wt_sub_dna = dna_counts_VSMC[dna_counts_VSMC["element"] == seq]
        wt_sub_dna = wt_sub_dna[cols_wt]

        cols_snp_rna = ["element"]+ list(snp_rna_cols_df['column'])
        cols_wt_rna = ["element"]+ list(wt_rna_cols_df['column'])        
        
        snp_sub_rna = rna_counts_VSMC[rna_counts_VSMC["element"] == seq]
        snp_sub_rna = snp_sub_rna[cols_snp_rna]
        wt_sub_rna = rna_counts_VSMC[rna_counts_VSMC["element"] == seq]
        wt_sub_rna = wt_sub_rna[cols_wt_rna]

        snp_rna_cols.extend(["%s___seq:SNP" % x for x in snp_sub_rna.columns if x != "element" ])
        wt_rna_cols.extend(["%s___seq:WT" % x for x in wt_sub_rna.columns if x != "element"])

        snp_dna_cols.extend(["%s___seq:SNP" % x for x in snp_sub_dna.columns if x != "element" ])
        wt_dna_cols.extend(["%s___seq:WT" % x for x in wt_sub_dna.columns if x != "element" ])

        snp_sub_dna.columns = snp_dna_cols
        wt_sub_dna.columns = wt_dna_cols
        snp_sub_rna.columns = snp_rna_cols
        wt_sub_rna.columns = wt_rna_cols

                # add comp_id to each df
        comp_id = "CONTROL:%s__SAMP_PAIR:%s" % ((i+1), (j+1))
        print(comp_id)
        snp_sub_dna["comp_id"] = comp_id
        wt_sub_dna["comp_id"] = comp_id
        snp_sub_rna["comp_id"] = comp_id
        wt_sub_rna["comp_id"] = comp_id

                # merge each df into 1
        snp_sub_dna.drop("element", axis=1, inplace=True)
        wt_sub_dna.drop("element", axis=1, inplace=True)
        snp_sub_rna.drop("element", axis=1, inplace=True)
        wt_sub_rna.drop("element", axis=1, inplace=True)

        sub_dna = snp_sub_dna.merge(wt_sub_dna, on="comp_id")
        sub_rna = snp_sub_rna.merge(wt_sub_rna, on="comp_id")

                # append
        neg_ctrls_dna = neg_ctrls_dna.append(sub_dna)
        neg_ctrls_rna = neg_ctrls_rna.append(sub_rna)
        
```

## 7. make annotation files

```python
all_rna.drop(['element___seq:WT'],axis=1, inplace=True)
all_dna.drop(['element___seq:WT'],axis=1, inplace=True)
all_rna.head()
```

```python
all_rna.drop(['tile_type___seq:WT'],axis=1, inplace=True)
all_dna.drop(['tile_type___seq:WT'],axis=1, inplace=True)
all_rna.head()
```

```python
all_dna_cols = all_dna.columns
all_rna_cols = all_rna.columns
```

```python
dna_col_ann = {}
rna_col_ann = {}

for cols, ann in zip([all_dna_cols, all_rna_cols], 
                     [dna_col_ann, rna_col_ann]):
    for col in cols:
        #print(col)
        if col == "comp_id":
            continue
        cond = col.split(":")[1].split("_")[0]
        barc = col.split(":")[2].split("_")[0]
        seq = col.split(":")[-1]
        ann[col] = {"condition": cond, "barcode": barc, "seq": seq}

dna_col_ann = pd.DataFrame.from_dict(dna_col_ann, orient="index")
rna_col_ann = pd.DataFrame.from_dict(rna_col_ann, orient="index")
rna_col_ann.sample(5)
```

```python
# merge w/ older annotations: first reset index
rna_col_ann.reset_index(inplace=True)

rna_col_ann["colname"] = rna_col_ann["index"]
rna_col_ann.head()
```

```python
# reset index
rna_col_ann.set_index("colname", inplace=True)
```

```python
rna_col_ann.index.name = None
rna_col_ann.sample(5)
```
