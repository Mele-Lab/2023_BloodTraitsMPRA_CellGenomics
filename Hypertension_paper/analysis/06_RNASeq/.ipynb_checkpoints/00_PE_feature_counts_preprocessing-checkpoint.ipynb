```python
rs3753326_33_1 = 'feature_counts/rs3753326_33_1/rs3753326_33_1.counts.txt'
rs3753326_33_2 = 'feature_counts/rs3753326_33_2/rs3753326_33_2.counts.txt'
rs3753326_50 = 'feature_counts/rs3753326_50/rs3753326_50.counts.txt'
rs3753326_40 = 'feature_counts/rs3753326_40/rs3753326_40.counts.txt'
rs3753326_74 = 'feature_counts/rs3753326_74/rs3753326_74.counts.txt'
rs3753326_70 = 'feature_counts/rs3753326_70/rs3753326_70.counts.txt'
rs3824754_58_1 = 'feature_counts/rs3824754_58_1/rs3824754_58_1.counts.txt'
rs3824754_58_2 = 'feature_counts/rs3824754_58_2/rs3824754_58_2.counts.txt'
rs3824754_37_1 = 'feature_counts/rs3824754_37_1/rs3824754_37_1.counts.txt'
rs3824754_37_2  = 'feature_counts/rs3824754_37_2/rs3824754_37_2.counts.txt'
rs4631439_30_1 = 'feature_counts/rs4631439_30_1/rs4631439_30_1.counts.txt'
rs4631439_30_2 = 'feature_counts/rs4631439_30_2/rs4631439_30_2.counts.txt'
rs4631439_60_1 = 'feature_counts/rs4631439_60_1/rs4631439_60_1.counts.txt'
rs4631439_60_2 = 'feature_counts/rs4631439_60_2/rs4631439_60_2.counts.txt'
rs4631439_100_1 = 'feature_counts/rs4631439_100_1/rs4631439_100_1.counts.txt'
rs4631439_100_2 = 'feature_counts/rs4631439_100_2/rs4631439_100_2.counts.txt'
rs6712094_80 = 'feature_counts/rs6712094_80/rs6712094_80.counts.txt'
rs6712094_86 = 'feature_counts/rs6712094_86/rs6712094_86.counts.txt'
WT_cl1 = 'feature_counts/WT_cl1/WT_cl1.counts.txt'
WT_cl2 = 'feature_counts/WT_cl2/WT_cl2.counts.txt'
```

## 1. import data 

```python
rs3753326_33_1_c = pd.read_table(rs3753326_33_1,skiprows=1)
rs3753326_33_2_c = pd.read_table(rs3753326_33_2,skiprows=1)
rs3753326_50_c = pd.read_table(rs3753326_50,skiprows=1)
rs3753326_40_c = pd.read_table(rs3753326_40,skiprows=1)
rs3753326_74_c = pd.read_table(rs3753326_74,skiprows=1)
rs3753326_70_c = pd.read_table(rs3753326_70,skiprows=1)

rs3824754_58_1_c = pd.read_table(rs3824754_58_1,skiprows=1)
rs3824754_58_2_c = pd.read_table(rs3824754_58_2,skiprows=1)
rs3824754_37_1_c = pd.read_table(rs3824754_37_1,skiprows=1)
rs3824754_37_2_c = pd.read_table(rs3824754_37_2,skiprows=1)

rs4631439_30_1_c = pd.read_table(rs4631439_30_1,skiprows=1)
rs4631439_30_2_c = pd.read_table(rs4631439_30_2,skiprows=1)
rs4631439_60_1_c = pd.read_table(rs4631439_60_1,skiprows=1)
rs4631439_60_2_c = pd.read_table(rs4631439_60_2,skiprows=1)
rs4631439_100_1_c = pd.read_table(rs4631439_100_1,skiprows=1)
rs4631439_100_2_c = pd.read_table(rs4631439_100_2,skiprows=1)

rs6712094_80_c = pd.read_table(rs6712094_80,skiprows=1)
rs6712094_86_c = pd.read_table(rs6712094_86,skiprows=1)
WT_cl1_c = pd.read_table(WT_cl1,skiprows=1)
WT_cl2_c = pd.read_table(WT_cl2,skiprows=1)
```

## 2. Heatmap comparing counts between samples 


```python
WT_cl2_c.iloc[:,[0,6]]
```

```python
rs3753326_33_1_counts = rs3753326_33_1_c.iloc[:,[0,6]]
rs3753326_33_2_counts = rs3753326_33_2_c.iloc[:,[0,6]]
rs3753326_50_counts = rs3753326_50_c.iloc[:,[0,6]]
rs3753326_40_counts = rs3753326_40_c.iloc[:,[0,6]]
rs3753326_74_counts = rs3753326_74_c.iloc[:,[0,6]]
rs3753326_70_counts = rs3753326_70_c.iloc[:,[0,6]]

rs3824754_58_1_counts = rs3824754_58_1_c.iloc[:,[0,6]]
rs3824754_58_2_counts = rs3824754_58_2_c.iloc[:,[0,6]]
rs3824754_37_1_counts = rs3824754_37_1_c.iloc[:,[0,6]]
rs3824754_37_2_counts = rs3824754_37_2_c.iloc[:,[0,6]]

rs4631439_30_1_counts = rs4631439_30_1_c.iloc[:,[0,6]]
rs4631439_30_2_counts = rs4631439_30_2_c.iloc[:,[0,6]]
rs4631439_60_1_counts = rs4631439_60_1_c.iloc[:,[0,6]]
rs4631439_60_2_counts = rs4631439_60_2_c.iloc[:,[0,6]]
rs4631439_100_1_counts = rs4631439_100_1_c.iloc[:,[0,6]]
rs4631439_100_2_counts = rs4631439_100_2_c.iloc[:,[0,6]]

rs6712094_80_counts = rs6712094_80_c.iloc[:,[0,6]]
rs6712094_86_counts = rs6712094_86_c.iloc[:,[0,6]]
WT_cl1_counts = WT_cl1_c.iloc[:,[0,6]]
WT_cl2_counts = WT_cl2_c.iloc[:,[0,6]]
```

```python
rs3753326_33_1_counts.columns = ['Geneid','rs3753326_33_1']
rs3753326_33_2_counts.columns = ['Geneid','rs3753326_33_2']
rs3753326_50_counts.columns = ['Geneid','rs3753326_50']
rs3753326_40_counts.columns = ['Geneid','rs3753326_40']
rs3753326_74_counts.columns = ['Geneid','rs3753326_74']
rs3753326_70_counts.columns = ['Geneid','rs3753326_70']

rs3824754_58_1_counts.columns = ['Geneid','rs3824754_58_1']
rs3824754_58_2_counts.columns = ['Geneid','rs3824754_58_2']
rs3824754_37_1_counts.columns = ['Geneid','rs3824754_37_1']
rs3824754_37_2_counts.columns = ['Geneid','rs3824754_37_2']

rs4631439_30_1_counts.columns = ['Geneid','rs4631439_30_1']
rs4631439_30_2_counts.columns = ['Geneid','rs4631439_30_2']
rs4631439_60_1_counts.columns = ['Geneid','rs4631439_60_1']
rs4631439_60_2_counts.columns = ['Geneid','rs4631439_60_2']
rs4631439_100_1_counts.columns = ['Geneid','rs4631439_100_1']
rs4631439_100_2_counts.columns = ['Geneid','rs4631439_100_2']

rs6712094_80_counts.columns = ['Geneid','rs6712094_80']
rs6712094_86_counts.columns = ['Geneid','rs6712094_86']
WT_cl1_counts.columns = ['Geneid','WT_cl1']
WT_cl2_counts.columns = ['Geneid','WT_cl2']
```

```python
all_counts = rs3753326_33_1_counts.merge(rs3753326_33_2_counts,how='outer').merge(rs3753326_50_counts,how='outer').merge(
    rs3753326_40_counts,how='outer').merge(rs3753326_74_counts,how='outer').merge(
    rs3753326_70_counts,how='outer').merge(rs3824754_58_1_counts,how='outer').merge(
    rs3824754_58_2_counts,how='outer').merge(rs3824754_37_1_counts,how='outer').merge(
    rs3824754_37_2_counts,how='outer').merge(rs3824754_73_counts,how='outer').merge(
    rs3824754_74_counts,how='outer').merge(rs3824754_73_2_counts,how='outer').merge(
    rs3824754_74_2_counts,how='outer').merge(rs4631439_30_1_counts,how='outer').merge(
    rs4631439_30_2_counts,how='outer').merge(rs4631439_60_1_counts,how='outer').merge(
    rs4631439_60_2_counts,how='outer').merge(rs4631439_100_1_counts,how='outer').merge(
    rs4631439_100_2_counts,how='outer').merge(rs6712094_80_counts,how='outer').merge(
    rs6712094_86_counts,how='outer').merge(WT_cl1_counts,how='outer').merge(
    WT_cl2_counts,how='outer')
```

```python
all_samples = all_counts.copy()
all_samples.drop("Geneid", axis=1, inplace=True)
all_samples = np.log10(all_samples+1)
all_samples_corr = all_samples.corr(method="spearman")
```

```python
cmap = sns.cubehelix_palette(as_cmap=True)
cg = sns.clustermap(all_samples_corr, figsize=(16,12), cmap=cmap, annot=True)
_ = plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
plt.suptitle("pearson correlation of replicates\nlog10+1 counts of genes")
plt.subplots_adjust(top=0.8)
cg.savefig("feature_counts/all_samples_corr_heatmap.spearman.PE.pdf", dpi="figure", transparent=True, bbox_inches="tight")
```
